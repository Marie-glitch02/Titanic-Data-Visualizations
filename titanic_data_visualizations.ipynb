```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme()

import warnings
warnings.filterwarnings("ignore")
```

```python
#Import the data set
titanic_data = pd.read_csv('titanic_train.csv')
```


## Visualizing Missing Data

```python
from matplotlib.colors import ListedColormap
cmap_dict = {0: '#FFFFFF', 1: '#ff2a00', 2: '#ff5500', 3: '#ff8000', 4: '#ffaa00', 5: '#ffd500'}
cmap = ListedColormap([cmap_dict[i] for i in range(6)])

sns.heatmap(titanic_data.isnull(), cmap=cmap, cbar=False)
```


## Survival Count Plot

```python
sns.countplot(x='Survived', data=titanic_data)
```

## Survival Count by Gender

```python
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
```

## Age Distribution Histogram


```python
plt.hist(titanic_data['Age'].dropna(), color='#ffd500')
```

## Age Distribution by Passenger Class


```python
sns.boxplot(x=titanic_data['Pclass'], y=titanic_data['Age'])
```
